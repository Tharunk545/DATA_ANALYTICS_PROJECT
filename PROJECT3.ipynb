```python
# Define a custom function to search for rows with 'score' greater than 1
def custom_search(row):
    return row['score'] > 1

# Apply the custom function to filter rows
result = data[data.apply(custom_search, axis=1)]
print(result.head)
```

#**Step 6: Data Visualisation**

```python
print(data.columns)

# To create a Dataframe object to hold the data in a tabular format
df = pd.DataFrame(data.head(20))
```

```python
# Generate a pie chart to visualize the distribution of values in the 'score' column
df['score'].value_counts().plot(kind='pie', autopct='%1.1f%%')

# Add a title to the pie chart indicating the variable being visualized
plt.title("score")

# Display the pie chart
plt.show()
```

```python
# Generate a bar chart to visualize the distribution of values in the 'country' column
data['country'].value_counts().plot(kind='bar', color='yellow')

# Add a label to the x-axis indicating the variable being visualized
plt.xlabel('Number of is country')

# Add a label to the y-axis indicating the count of each value
plt.ylabel('Count')

# Add a title to the bar chart indicating the purpose of the visualization
plt.title('Distribution of Number of country')

# Display the bar chart
plt.show()
```

```python
# Generate a horizontal bar chart to visualize the distribution of values in the 'score' column
data['score'].value_counts().plot(kind='barh', color='black')

# Add a label to the x-axis indicating the count of each value
plt.xlabel('Count')

# Add a label to the y-axis indicating the variable being visualized
plt.ylabel('score Level')

# Add a title to the horizontal bar chart indicating the purpose of the visualization
plt.title('Distribution of score Levels')

# Display the horizontal bar chart
plt.show()
```

```python
# Convert the 'score' column to datetime format if it's not already in datetime format
data['score'] = pd.to_datetime(data['score'])

# Sort the DataFrame by the 'score' column to ensure the data is plotted in chronological order
data.sort_values(by='score', inplace=True)

# Plot a line chart to visualize the trend of 'Income' over time
plt.plot(data['score'], data['country'], color='blue', marker='o', linestyle='-')

# Add a label to the x-axis indicating the date of customer registration
plt.xlabel('score')

# Add a label to the y-axis indicating the income
plt.ylabel('country')

# Add a title to the line chart indicating the purpose of the visualization
plt.title('score of individual country')

# Rotate x-axis labels by 45 degrees for better readability
plt.xticks(rotation=45)

# Add grid lines to the plot for better visualization
plt.grid(True)

# Display the line chart
plt.show()
```

```python
# Generate a histogram to visualize the distribution of values in the 'score' column
plt.hist(data['score'], bins=20, color='green', edgecolor='black')

# Add a label to the x-axis indicating the variable being visualized
plt.xlabel('score')

# Add a label to the y-axis indicating the frequency of observations
plt.ylabel('Frequency')

# Add a title to the histogram indicating the purpose of the visualization
plt.title('Distribution of score')

# Add grid lines to the histogram for better visualization
plt.grid(True)

# Display the histogram
plt.show()
```

```python
# Generate a scatter plot to visualize the relationship between 'Income' and 'NumWebPurchases'
plt.scatter(data['score'], data['version'], color='blue', alpha=0.5)

# Add a label to the x-axis indicating the variable being visualized
plt.xlabel('score')

# Add a label to the y-axis indicating the variable being visualized
plt.ylabel('versions')

# Add a title to the scatter plot indicating the purpose of the visualization
plt.title('Scatter Plot: score vs. Number of versions')

# Add grid lines to the scatter plot for better visualization
plt.grid(True)

# Display the scatter plot
plt.show()
```
